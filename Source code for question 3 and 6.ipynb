Analytical Computation

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#Defining the function for u and v
u=lambda x,y:(x**2-y**2)
v=lambda x,y:(-1*x*y)
```

```python
C=np.linspace(-1*2/3,2/3,100)
```

```python
plt.figure(figsize=(6,6))
for c in C:
    Y=np.linspace(-1,1,10000)
    c_plot=[c]*10000
    c_plot=np.array(c_plot)
    X=(c_plot/Y-Y**2/3)**0.5
    X1=-1*(c_plot/Y-Y**2/3)**0.5
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    plt.plot(X,Y)
    plt.plot(X1,Y)
    plt.title('Analytical Plot')
```

Numerical Computation

```python
X=np.linspace(-1,0,100)
```

```python
#function for a streamline
def mystreamline(X,y0):
    Y=[]
    x0=-1
    for x in X[:]:
        #Integration using trapezoidal rule
        y=((-2*x0*y0)/(x0**2-y0**2))*x+((3*(x0**2)*y0-y0**3)/(x0**2-y0**2))
        #Finding y of point2 from tangent equation at point1
        Y.append(y0)
        x0=x
        y0=y
    Y=np.array(Y)
    return Y
```
